## a. Solucionar sistema.

```python
#Se definen las constantes
a = 0.24250109 
b = -0.0607718825
c = 0.93240737 
d = 0.86473146
#Se definen las posiciones iniciales.
r1init=np.array([a, b])
r2init=np.array([-a, -b])
r3init=np.array([0., 0.])
#Se definen las velocidades iniciales.
v1init=np.array([c, d])
v2init=v1init
v3init=-2.*v1init
#Se definen los vectores de condiciones iniciales de la forma (x, y, vx, vy).
z1=np.array([r1init[0], r1init[1], v1init[0], v1init[1]])
z2=np.array([r2init[0], r2init[1], v2init[0], v2init[1]])
z3=np.array([r3init[0], r3init[1], v3init[0], v3init[1]])
#Se define la constante gravitacional.
G=1.
```

```python
#Se define una función para la aceleración sobre los cuerpos que recibe como parámetros los vectores de estado de éstos.
def f(Y1, Y2, Y3):
    #Se calcula la fuerza dada en cada cuerpo a causa de los otros dos para las componentes x e y .
    gravForce1x = (Y1[0]-Y2[0])*(-G/(np.sqrt((Y1[0]-Y2[0])**2+(Y1[1]-Y2[1])**2)**3)) - (Y1[0]-Y3[0])*(G/(np.sqrt((Y1[0]-Y3[0])**2+(Y1[1]-Y3[1])**2)**3))
    gravForce2x = (Y2[0]-Y1[0])*(-G/(np.sqrt((Y2[0]-Y1[0])**2+(Y2[1]-Y1[1])**2)**3)) - (Y2[0]-Y3[0])*(G/(np.sqrt((Y2[0]-Y3[0])**2+(Y2[1]-Y3[1])**2)**3))
    gravForce3x = (Y3[0]-Y1[0])*(-G/(np.sqrt((Y3[0]-Y1[0])**2+(Y3[1]-Y1[1])**2)**3)) - (Y3[0]-Y2[0])*(G/(np.sqrt((Y3[0]-Y2[0])**2+(Y3[1]-Y2[1])**2)**3))
    
    gravForce1y = (Y1[1]-Y2[1])*(-G/(np.sqrt((Y1[0]-Y2[0])**2+(Y1[1]-Y2[1])**2)**3)) - (Y1[1]-Y3[1])*(G/(np.sqrt((Y1[0]-Y3[0])**2+(Y1[1]-Y3[1])**2)**3))
    gravForce2y = (Y2[1]-Y1[1])*(-G/(np.sqrt((Y2[0]-Y1[0])**2+(Y2[1]-Y1[1])**2)**3)) - (Y2[1]-Y3[1])*(G/(np.sqrt((Y2[0]-Y3[0])**2+(Y2[1]-Y3[1])**2)**3))
    gravForce3y = (Y3[1]-Y2[1])*(-G/(np.sqrt((Y3[0]-Y2[0])**2+(Y3[1]-Y2[1])**2)**3)) - (Y3[1]-Y1[1])*(G/(np.sqrt((Y3[0]-Y1[0])**2+(Y3[1]-Y1[1])**2)**3))
    
    #Se devuelven tres vectores de estado actualizados. Para los cuerpos 1, 2 y 3 respectivamente.
    return np.array([Y1[2], Y1[3], gravForce1x,gravForce1y]), np.array([Y2[2], Y2[3], gravForce2x,gravForce2y]), np.array([Y3[2], Y3[3], gravForce3x,gravForce3y])
```

```python
#Se define la función que toma la función anterior y aproxima sus valores en t + dt.
def one_Kutta4_step3B():
    global t
    global Deltat
    global Y1, Y2, Y3
    #Se evalúa la función en su estado inicial para obtener K1.
    K1 = f(Y1, Y2, Y3)
    #Se aplica la regla de RK4 en los tres vectores de estado para actualizarlos también.
    K2 = f(Y1 + Deltat*(1./2.*K1[0]), Y2 + Deltat*(1./2.*K1[1]), Y3 + Deltat*(1./2.*K1[2]))
    K3 = f(Y1 + Deltat*(1./2.*K2[0]), Y2 + Deltat*(1./2.*K2[1]), Y3 + Deltat*(1./2.*K2[2]))
    K4 = f(Y1 + Deltat*K3[0], Y2 + Deltat*K3[1], Y3 + Deltat*K3[2])
    #Se modifican los vectores.
    Y1+= Deltat * (1/6.*K1[0] + 1/3. * K2[0] + 1/3.*K3[0] + 1/6.*K4[0])
    Y2+= Deltat * (1/6.*K1[1] + 1/3. * K2[1] + 1/3.*K3[1] + 1/6.*K4[1])
    Y3+= Deltat * (1/6.*K1[2] + 1/3. * K2[2] + 1/3.*K3[2] + 1/6.*K4[2])
    t+= Deltat
```

```python
#Se definen los vectores de condiciones iniciales.
z1=np.array([r1init[0], r1init[1], v1init[0], v1init[1]])
z2=np.array([r2init[0], r2init[1], v2init[0], v2init[1]])
z3=np.array([r3init[0], r3init[1], v3init[0], v3init[1]])
#Se define el numero de iteraciones.
numIter=100
#Se inicializan los vectores de estado con las condiciones iniciales.
Y1=np.copy(z1)
Y2=np.copy(z2)
Y3=np.copy(z3)
#Se da el intervalo de tiempo.
tmax=0.5
t=0.
#Se da el timestep.
Deltat=tmax/numIter
#Se crea un arreglo que guarde la información de cada cuerpo en cada tiempo y se inicializa con las condiciones iniciales.
history1=np.array([[t,Y1[0],Y1[1],Y1[2],Y1[3]]])
history2=np.array([[t,Y2[0],Y2[1],Y2[2],Y2[3]]])
history3=np.array([[t,Y3[0],Y3[1],Y3[2],Y3[3]]])
#Se aplica Rnge-Kutta numIter veces, actualizando el arreglo con los estados.
for __ in range(numIter):    
    one_Kutta4_step3B()
    history1=np.append(history1,[[t,Y1[0],Y1[1],Y1[2],Y1[3]]],axis=0)
    history2=np.append(history2,[[t,Y2[0],Y2[1],Y2[2],Y2[3]]],axis=0)
    history3=np.append(history3,[[t,Y3[0],Y3[1],Y3[2],Y3[3]]],axis=0)
#Se grafica la trayectoria de las masas en el tiempo dado.
Trayectorias, ax=plt.subplots(3, 1, figsize=(10, 7), sharex=True)
ax[0].plot(history1[:,1],history1[:,2], c="red")
ax[1].plot(history2[:,1],history2[:,2], c="blue")
ax[2].plot(history3[:,1],history3[:,2], c="green")

ax[0].scatter(history1[-1,1],history1[-1,2], c="red")
ax[0].set_title("Trayectoria de $m_1$ en t=[0, 0.5]")
ax[2].set_xlabel("$x$", fontsize=15)
ax[1].set_ylabel("$y$", fontsize=15)

ax[1].scatter(history2[-1,1],history2[-1,2], c="blue")
ax[1].set_title("Trayectoria de $m_2$ en t=[0, 0.5]")
ax[2].scatter(history3[-1,1],history3[-1,2], c="green")
ax[2].set_title("Trayectoria de $m_3$ en t=[0, 0.5]")
ax[2].set_ylim(-.1, .1)
plt.tight_layout()
plt.show()
```

```python
#Es necesario volver a solucionar la ecuación pero para un tiempo mayor, de forma que se asegure que hay, por lo menos un periodo en él.
#Se definen los vectores de condiciones iniciales.
z1=np.array([r1init[0], r1init[1], v1init[0], v1init[1]])
z2=np.array([r2init[0], r2init[1], v2init[0], v2init[1]])
z3=np.array([r3init[0], r3init[1], v3init[0], v3init[1]])
#Se define el numero de iteraciones. Más alto para aumentar la precisión.
numIter=10000
#Se inicializan los vectores de estado con las condiciones iniciales.
Y1=np.copy(z1)
Y2=np.copy(z2)
Y3=np.copy(z3)
#Se da el intervalo de tiempo.
tmax=1.1
t=0.
#Se da el timestep. Se pone un número 'extraño' en el denominador para obligar al programa a usar más de dos cifras para los tiempos
#y así calcular con mayor precisión el periodo.
Deltat=tmax/numIter/1.2
#Se crea un arreglo que guarde la información de cada cuerpo en cada tiempo y se inicializa con las condiciones iniciales.
history1_p=np.array([[t,Y1[0],Y1[1],Y1[2],Y1[3]]])
history2_p=np.array([[t,Y2[0],Y2[1],Y2[2],Y2[3]]])
history3_p=np.array([[t,Y3[0],Y3[1],Y3[2],Y3[3]]])
#Se aplica Runge-Kutta numIter veces, actualizando el arreglo con los estados.
for __ in range(numIter):    
    one_Kutta4_step3B()
    history1_p=np.append(history1_p,[[t,Y1[0],Y1[1],Y1[2],Y1[3]]],axis=0)
    history2_p=np.append(history2_p,[[t,Y2[0],Y2[1],Y2[2],Y2[3]]],axis=0)
    history3_p=np.append(history3_p,[[t,Y3[0],Y3[1],Y3[2],Y3[3]]],axis=0)

    
#Se extrae el arreglo con los tiempos.
times=history1_p[:, 0]
#Se extrae el arreglo con las coordenadas espaciales iniciales.
x0=history1_p[0, 1]
y0=history1_p[0, 2]


#Se hace un arreglo que tome las distancias de cada posición en t<0, al punto inicial.
pos1=np.sqrt((history1_p[:, 1]-x0)**2+(history1_p[:, 2]-y0)**2)
#Se busca el primer mínimo de la función ignorando el 0 inicial.
minsDiff0=np.min(pos1[1:])
#Se Extrae del arreglo de tiampos el tiempo en el que el anterior vector tiene un mínimo.
Period=times[np.where(pos1==minsDiff0)]

h = (b-a)/(numIter-1) # Este es el paso del arreglo x
yforward = np.roll(pos1,-1) # toma el arreglo y y lo mueve una posición hacia la izquierda
ybackward = np.roll(pos1,1) # y esto una posición a la derecha
# y finalmente usando diferencias centrales.
derivPos = (yforward-ybackward)/(2.*h)
# La primera y la última no tienen significado
derivPos=derivPos[1:-1]
signos=np.sign(derivPos)
cambio = ((np.roll(signos, 1) - signos) != 0).astype(int)
P3=times[cambio==1]
#Se toma el segundo elemento dado que el primero corresponde al máximo de la función.

fig, ax=plt.subplots(2, 1, figsize=(10, 7), sharex=True)
ax[0].grid()
ax[1].grid()
ax[0].plot(times, pos1)
ax[0].set_title("$R(t)$")
ax[1].plot(times[1:-1], derivPos)
ax[1].set_title("$dR(t)/dt$")
ax[1].set_xlabel("$t$")

plt.show()
print "El periodo de de acuerdo al mínimo de la función de distancias es de:", Period[0]
print "El periodo de acuerdo con la prueba del cambio de signo en la derivada es: ", P3[1]

```

## a. Se busca el parámetro *a*

```python
def g(H1, H2, H3, H4):
    #Se calcula la fuerza dada en cada cuerpo a causa de los otros tres para las componentes x e y .
    gravForce1x = (H1[0]-H2[0])*(-G/(np.sqrt((H1[0]-H2[0])**2+(H1[1]-H2[1])**2)**3)) - (H1[0]-H3[0])*(G/(np.sqrt((H1[0]-H3[0])**2+(H1[1]-H3[1])**2)**3)) - (H1[0]-H4[0])*(G/(np.sqrt((H1[0]-H4[0])**2+(H1[1]-H4[1])**2)**3))
    gravForce2x = (H2[0]-H1[0])*(-G/(np.sqrt((H2[0]-H1[0])**2+(H2[1]-H1[1])**2)**3)) - (H2[0]-H3[0])*(G/(np.sqrt((H2[0]-H3[0])**2+(H2[1]-H3[1])**2)**3)) - (H2[0]-H4[0])*(G/(np.sqrt((H2[0]-H4[0])**2+(H2[1]-H4[1])**2)**3))
    gravForce3x = (H3[0]-H2[0])*(-G/(np.sqrt((H3[0]-H2[0])**2+(H3[1]-H2[1])**2)**3)) - (H3[0]-H1[0])*(G/(np.sqrt((H3[0]-H1[0])**2+(H3[1]-H1[1])**2)**3)) - (H3[0]-H4[0])*(G/(np.sqrt((H3[0]-H4[0])**2+(H3[1]-H4[1])**2)**3))
    gravForce4x = (H4[0]-H1[0])*(-G/(np.sqrt((H4[0]-H1[0])**2+(H4[1]-H1[1])**2)**3)) - (H4[0]-H2[0])*(G/(np.sqrt((H4[0]-H2[0])**2+(H4[1]-H2[1])**2)**3)) - (H4[0]-H3[0])*(G/(np.sqrt((H4[0]-H3[0])**2+(H4[1]-H3[1])**2)**3))
    
    gravForce1y = (H1[1]-H2[1])*(-G/(np.sqrt((H1[0]-H2[0])**2+(H1[1]-H2[1])**2)**3)) - (H1[1]-H3[1])*(G/(np.sqrt((H1[0]-H3[0])**2+(H1[1]-H3[1])**2)**3)) - (H1[1]-H4[1])*(G/(np.sqrt((H1[0]-H4[0])**2+(H1[1]-H4[1])**2)**3))
    gravForce2y = (H2[1]-H1[1])*(-G/(np.sqrt((H2[0]-H1[0])**2+(H2[1]-H1[1])**2)**3)) - (H2[1]-H3[1])*(G/(np.sqrt((H2[0]-H3[0])**2+(H2[1]-H3[1])**2)**3)) - (H2[1]-H4[1])*(G/(np.sqrt((H2[0]-H4[0])**2+(H2[1]-H4[1])**2)**3))
    gravForce3y = (H3[1]-H2[1])*(-G/(np.sqrt((H3[0]-H2[0])**2+(H3[1]-H2[1])**2)**3)) - (H3[1]-H1[1])*(G/(np.sqrt((H3[0]-H1[0])**2+(H3[1]-H1[1])**2)**3)) - (H3[1]-H4[1])*(G/(np.sqrt((H3[0]-H4[0])**2+(H3[1]-H4[1])**2)**3))
    gravForce4y = (H4[1]-H1[1])*(-G/(np.sqrt((H4[0]-H1[0])**2+(H4[1]-H1[1])**2)**3)) - (H4[1]-H2[1])*(G/(np.sqrt((H4[0]-H2[0])**2+(H4[1]-H2[1])**2)**3)) - (H4[1]-H3[1])*(G/(np.sqrt((H4[0]-H3[0])**2+(H4[1]-H3[1])**2)**3))
    
    #Se devuelven cuatro vectores de estado actualizados. Para los cuerpos 1, 2, 3 y 4 respectivamente.
    
    return np.array([H1[2], H1[3], gravForce1x,gravForce1y]), np.array([H2[2], H2[3], gravForce2x,gravForce2y]), np.array([H3[2], H3[3], gravForce3x,gravForce3y]), np.array([H4[2], H4[3], gravForce4x,gravForce4y])
```

```python
def one_Kutta4_step4B():
    global t
    global Deltat
    global H1, H2, H3, H4
    #Se evalúa la función en su estado inicial para obtener K1.
    K1 = g(H1, H2, H3, H4)
    #Se aplica la regla de RK4 en los tres vectores de estado para actualizarlos también.
    K2 = g(H1 + Deltat*(1./2.*K1[0]), H2 + Deltat*(1./2.*K1[1]), H3 + Deltat*(1./2.*K1[2]), H4 + Deltat*(1./2.*K1[3]))
    K3 = g(H1 + Deltat*(1./2.*K2[0]), H2 + Deltat*(1./2.*K2[1]), H3 + Deltat*(1./2.*K2[2]), H4 + Deltat*(1./2.*K2[3]))
    K4 = g(H1 + Deltat*K3[0], H2 + Deltat*K3[1], H3 + Deltat*K3[2], H4 + Deltat*K3[3])
    #Se modifican los vectores.
    H1+= Deltat * (1/6.*K1[0] + 1/3. * K2[0] + 1/3.*K3[0] + 1/6.*K4[0])
    H2+= Deltat * (1/6.*K1[1] + 1/3. * K2[1] + 1/3.*K3[1] + 1/6.*K4[1])
    H3+= Deltat * (1/6.*K1[2] + 1/3. * K2[2] + 1/3.*K3[2] + 1/6.*K4[2])
    H4+= Deltat * (1/6.*K1[3] + 1/3. * K2[3] + 1/3.*K3[3] + 1/6.*K4[3])
    t+= Deltat
```

```python
#Se crea un arreglo de posibles valores de a entre 0.70 y 0.78. Luego se modificaron al averiguar que se encontraba alrededor de 0.760
poss_a=np.linspace(0.7601, 0.761, 200)
#Se hace un ciclo que solucione la ecuación para cada valor de a en poss_a
for i in range(len(poss_a)):
    #Se define el valor de la incognita
    unk_a=poss_a[i]
    #Se establecen las condiciones iniciales.
    z4_1=np.array([0.384277200514, 0., 0., unk_a])
    z4_2=np.array([-0.0156823005697, -0.13966430504, -2.01155925929, -1.19817066623])
    z4_3=np.array([-0.352912599375, 0.,0., 1.63619158614])
    z4_4=np.array([-0.0156823005697, 0.13966430504, 2.01155925929, -1.19817066623])
    chosenstep=one_Kutta4_step4B
    #Las iteraciones de la solución.
    numIter=10000
    #Se inicializan los vectores de estado con las condiciones iniciales.
    H1=z4_1
    H2=z4_2
    H3=z4_3
    H4=z4_4
    #Se elige un tiempo máximo lo suficientemente largo para que contenga al menos un periodo, se estima.
    tmax=2.
    t=0.
    #Se define el 'timestep'.
    Deltat=tmax/numIter/1.4
    #Se crean los arreglos con los estados y se inicializan con las condiciones iniciales.
    history4_1=np.array([[t,H1[0],H1[1],H1[2],H1[3]]])
    history4_2=np.array([[t,H2[0],H2[1],H2[2],H2[3]]])
    history4_3=np.array([[t,H3[0],H3[1],H3[2],H3[3]]])
    history4_4=np.array([[t,H4[0],H4[1],H4[2],H4[3]]])
    #Se aplica el algoritmo de Runge-Kutta numIter veces.
    for __ in range(numIter):    
        chosenstep()
        history4_1=np.append(history4_1,[[t,H1[0],H1[1],H1[2],H1[3]]],axis=0)
        history4_2=np.append(history4_2,[[t,H2[0],H2[1],H2[2],H2[3]]],axis=0)
        history4_3=np.append(history4_3,[[t,H3[0],H3[1],H3[2],H3[3]]],axis=0)
        history4_4=np.append(history4_4,[[t,H4[0],H4[1],H4[2],H4[3]]],axis=0)
    #Se hace un arreglo con los tiempos.
    times4_p=history4_1[:, 0]
    #Se crean los arreglos con las coordenadas del inicio.
    x04=history4_1[0, 1]
    y04=history4_1[0, 2]
    #Se crea el vector de distancias respecto al punto inicial de la trayectoria.
    pos4_1=np.sqrt((history4_1[:, 1]-x04)**2+(history4_1[:, 2]-y04)**2)
    #Se estima numéricamente la derivada.    
    h = (b-a)/(numIter-1)
    yforward = np.roll(pos4_1,-1)
    ybackward = np.roll(pos4_1,1)
    dy3 = (yforward-ybackward)/(2.*h)
    dy3=dy3[1:-1]
    #Se crea un vector de signos.
    asign=np.sign(dy3)
    #Se revisa donde cambian los signos.
    signchange = ((np.roll(asign, 1) - asign) != 0).astype(int)
    #Se obtiene el valor de t en el cual la derivada de pos4_1 cambia de signo.
    P=times4_p[signchange==1]
    #Se calcula el mínimo del arreglo, ignorando el cero inicial.
    minsDiff04=np.amin(pos4_1[1:])
    #Se define el periodo.
    Period4=times4_p[np.where(pos4_1==minsDiff04)]
    #Se hace un condicional que retorne el valor de a para el cual el periodo es 1. 
    if (np.round(P[1], 4) == 1.):
        the_a=unk_a
        print "a=", unk_a,"=",  np.round(unk_a, 5)
        break
print "Hecho" 
```

```python
#Se define una función que calcule y retorne la energía cinética de un cuerpo.

def KE(vx, vy):
    spd=np.sqrt(vx**2 + vy**2)
    KinE=spd**2/2.
    return KinE

#Se definen para cada cuerpo las velocidades en x e y, se calcula la energía
#cinética de cada uno.
vx_1=history4_1[:, 3]
vy_1=history4_1[:, 4]
KE_1=KE(vx_1, vy_1)

vx_2=history4_2[:, 3]
vy_2=history4_2[:, 4]
KE_2=KE(vx_2, vy_2)

vx_3=history4_3[:, 3]
vy_3=history4_3[:, 4]
KE_3=KE(vx_3, vy_3)

vx_4=history4_4[:, 3]
vy_4=history4_4[:, 4]
KE_4=KE(vx_4, vy_4)

#Se Obtiene el total de energía cunética.
KTot=KE_1+KE_2+KE_3+KE_4

#Se define una función que calcule y retorne la energía potencial.
def PE(x_a, x_b, y_a, y_b):
    PotE = -1./np.sqrt((x_b-x_a)**2 + (y_b - y_a)**2)
    return PotE

#Se extraen los vectores de coordenadas x e y de cada cuerpo.
x_1=history4_1[:, 1]
y_1=history4_1[:, 2]

x_2=history4_2[:, 1]
y_2=history4_2[:, 2]

x_3=history4_3[:, 1]
y_3=history4_3[:, 2]

x_4=history4_4[:, 1]
y_4=history4_4[:, 2]

#Se calcula la energía potencial entre cada combinación posible de cuerpos.
PE_12=PE(x_1, x_2, y_1, y_2)
PE_13=PE(x_1, x_3, y_1, y_3)
PE_14=PE(x_1, x_4, y_1, y_4)
PE_23=PE(x_2, x_3, y_2, y_3)
PE_24=PE(x_2, x_4, y_2, y_4)
PE_34=PE(x_3, x_4, y_3, y_4)

#Se suma la energía potencial de cada par de cuerpos y se obtiene la total.
PTot=PE_12+PE_13+PE_14+PE_23+PE_24+PE_34
#Se calcula la energía total.
ETot=KTot+PTot
#Se hace la gráfica de energias.
EnergyPlot, ax=plt.subplots(1, 1, figsize=(10, 7))
ax.set_title("$E$ vs. $t$")
ax.plot(times4, KTot, label="Kinetic Energy")
ax.plot(times4, PTot, label="Potential Energy")
ax.plot(times4, ETot, label="Total Energy")
ax.set_xlim(0, 1)
ax.set_xlabel("$t$")
ax.set_ylim(-20, 15)
ax.set_ylabel("$E$")
plt.legend()
plt.show(EnergyPlot)
```
