```python
#%pylab inline
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import stats
import csv
from scipy.optimize import curve_fit
import math
import scipy.misc
from scipy.stats import linregress
from scipy import interpolate
from scipy.fftpack import *
import scipy.signal
```

```python
#El lienzo
sq=np.zeros([1024, 1024])#, dtype=float)
#Pupil function
sq[512-32/2:512+32/2, 512-32/2:512+32/2]=1.
#La norma de la transformada
sqFFT=np.abs(fft2(sq))
#Se aplica la función fftshift
sq_fhf=scipy.fftpack.fftshift(sqFFT)
#Logaritmo del cuadrado de la norma mas uno (evitando el error log(0))
sq_fhf=np.log((sq_fhf+1)**2)
#Se muestran las imágenes
sqPlot, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].imshow(sq, cmap="Greys_r")
ax[1].imshow(sq_fhf[512-100:512+100, 512-100:512+100])
plt.title("Patron de interferencia de una apertura cuadrada\n (log resaltado)")
```

```python
#El lienzo.
dslit=np.zeros([1024, 1024], dtype=float)
#Pupil function.
dslit[512-100:512+100, 460-10:460+10]=1
dslit[512-100:512+100, 564-10:564+10]=1
#La norma de la transformada de Fourier.
dslitFFT=np.abs(fft2(dslit))
#La función fftshift.
dslit_fhf=scipy.fftpack.fftshift(dslitFFT)
#Se crea la imagen.
dSlit, ax = plt.subplots(1, 2, figsize=(15, 5))
ax[0].imshow(dslit, cmap="Greys_r")
ax[1].imshow(dslit_fhf[512-128/2:512+128/2, 512-128/2:512+128/2])
plt.title("Patron de difraccion de una doble rendija")
```

```python
#El lienzo
rnd=np.zeros([1024, 1024])#, dtype=float)
#Se fijan las coordenadas del centro.
cx,cy=512,512
#Se fija el radio.
cr=0.01*1024
#Se crean arreglos de indices.
I,J=np.meshgrid(range(1024),range(1024))
#Se calcula la distancia del centro a cada punto.
dist=np.sqrt((I-cx)**2+(J-cy)**2)
#Se reemplaza el valor por 1 en aquellos puntos donde la distancia sea menor al radio
rnd[dist<cr]=1
#Se realiza la transformada de Fourier y se toma la norma.
rndFFT=np.abs(fft2(rnd))
#Se aplica la función fftshift.
rnd_fhf=scipy.fftpack.fftshift(rndFFT)
#Se toma el cuadrado de la norma de la fila central.
i=np.abs(rnd_fhf[512])**2
#Se toma el cuadrado de la norma del valor central como I_0 para la intensidad máxima.
i0=np.abs(rnd_fhf[512, 512])**2
#Se calcula el array de valores de intensidad normalizada a la intensidad máxima.
ioveri0=i/i0
#Se crea el arreglo de abscisas co radio=0.01u.
x=2*np.pi*(0.01)/1.*np.linspace(-512,512,1024)
#Se crea la imagen.
rndPlot=plt.figure(figsize=(15, 10))
#Pupil function plot.
ax0 = plt.subplot2grid((2,2), (1, 0), colspan=1)
ax0.imshow(rnd, cmap="Greys_r")                    
#Patrón de difracción.
ax1 = plt.subplot2grid((2, 2), (1, 1))
ax1.imshow(np.log((rnd_fhf+1)**2))
plt.title("Patron de interferencia de una apertura circular (Patron de Airy)\n log resaltado")
#I vs. x
ax3 = plt.subplot2grid((2,2), (0, 0), colspan=2)
ax3.set_ylabel("$I/I_0$", fontsize=15)
ax3.set_xlabel("$2\pi Rj/L$", fontsize=15)
ax3.set_xlim(0, 10)
ax3.set_ylim(0, 1)
ax3.scatter(x, ioveri0)
rndPlot.tight_layout()
```

```python
#Se hace una interpolación tipo splice cúbico.
sCubic=scipy.interpolate.interp1d(x, ioveri0, kind="cubic")
```

```python
#Se crea el arreglo con elementos del 0 al 12 para aplicar la interpolación obtenida.
t=np.linspace(0, 12, 100)
#Se hace una figura para facilidad de visualización.
plt.plot(t, sCubic(t))
#Se aplica la interpolación al array de 0 a 12, creando el arreglo de ordenadas de la interpolación.
maxminvals=sCubic(t)
#Se extraen las posiciones de los máximos locales en el arreglo recién obtenido.
maxpos=scipy.signal.argrelextrema(maxminvals, np.greater)[0]
#Se extraen las posiciones de los mínimos locales en el mismo arreglo.
minpos=scipy.signal.argrelextrema(maxminvals, np.less)[0]
#Se crean las listas correspondientes a los ordenadas de los máximos y mínimos. Al aplicar la función argelextrema() se ve que no
#se obtiene el máximo absoluto de la función ni su posición, por lo cual se incluye en la lista desde el inico, basta con llamar el
#primer valor del arreglo de ordenadas de la interpolación.
maxima=[maxminvals[0]]
minima=[]
#Se crean las listas correspondientes a las abscisas de los máximos y mínimos. Se inicia la de máximos con el valor 0 o t[0].
xmax=[t[0]]
xmin=[]
#Se incluyen el resto de valores de máximos y mínimos en las listas correspondientes.
for i in range(len(maxpos)):
    maxima.append(maxminvals[maxpos[i]])
    xmax.append(t[maxpos[i]])
    minima.append(maxminvals[minpos[i]])
    xmin.append(t[minpos[i]])
#Se imprimen las respuestas.
print "Los máximos son: ", maxima
print "Los mínimos son: ", minima
print "Los máximos están en: ", xmax
print "Los mínimos están en: ", xmin
#Se ubican en la imagen los puntos correspondientes a los extremos.
plt.scatter(xmax, maxima)
plt.scatter(xmin, minima)
```
