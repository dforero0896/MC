## a. Función `lagrange`

```python
def lagrange(n):
    #Se crea una lista que conendrá los fn y los tn.
    f=[]
    tn=[]
    #Se definen t y h como símbolos.
    t=var("t")
    h=var("h")
    for i in range(n):
        #Se agregan a las listas el número de fn y tn correspondientes al orden y se convierten en símbolos.
        f.append(str("f%s" % i))
        f[i]=var(f[i])
        tn.append(str("t%s" % i))
        tn[i]=var(tn[i])
    #Se define una lista que contendrá elementos correspondientes a cada término del polinomio de orden n.
    polynomial=[]
    #Se itera sobre cada elemento en la lista.
    for i in range(n):
        #Se inicializa el término a agregar en 1.
        term=1
        #Se itera sobre el i-ésimo elemento en la lista para agregar los factores necesarios a éste según el orden del polinomio.
        for j in range(n):
            #Se exige que, para agregar un factor, se tomen los j diferentes a i, que corresponde al término que se está manejando.
            if i!=j:
                term *= (t-tn[j])/(tn[i]-tn[j])
        #Se agrega el i-ésimo término a la lista-polinomio.
        polynomial.append(f[i]*term) 
    #Se convierte la lsita en una expresión (polinomio).
    p=sum(polynomial)
    #Se regresan los f, los tn y el polinomio por razones prácticas en la implementación de la función 'adams(m)'.
    return f, tn, p
```

```python
def adams(m):
    #Se recuperan los valores de f, tn y p dados por la función 'lagrange(n)'.
    f, tn, p = lagrange(m)
    #Se itera sobre cada elemento en la lista de tn, sustituyendo tn[i]=tn[0]-i*h en el polinomio p usando la función subs(aSustituir, sustituido).
    for i in range(1, m):
        p = p.subs(tn[i],tn[0]-i*h)
    #Se devuelve la integral respecto a t de la función p en el intervalo [tn[0], tn[0]+h].
    return simplify(integrate(p, (t, tn[0], tn[0]+h )))
```
