```python
L=np.linspace(-pi, 100, 1000)
```

```python
def FuncSen(x, orden):
    n=1
    total=0  
    while n<orden:
        a = (-1)**(n-1)*((x**(2*n-1))/(math.factorial(2*n-1)))
        total+=a
        n += 1
    return total

Funcsen, ax=subplots(1, 1, figsize=(20, 14))
ax.set_xlim(-pi, 10*pi)
ax.set_ylim(-2, 2)
for i in range(50):
   ax.plot(L, FuncSen(L, i)) 



#ax.plot(L, sin(L))
ax.legend()
```
