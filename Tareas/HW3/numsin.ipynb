```python
#Se define una función que calcula la serie de Maclaurin para sin(x) sin usar
#la función factorial.
def FuncSen(x, orden):
    term=x
    total=term
    for n in range(1, orden):
        term*=-x**2/((2*n+1)*(2*n))
        total+=term
    return total
```
