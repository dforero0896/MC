```python
L=np.linspace(-pi, 100, 1000)
```

```python
def FuncSenFactorial(x, orden):
    n=1
    total=0  
    while n<orden:
        a = (-1)**(n-1)*((x**(2*n-1))/(math.factorial(2*n-1)))
        total+=a
        n += 1
    return total

Funcsenfac, ax=subplots(1, 1, figsize=(10, 7))
ax.set_xlim(-pi, 10*pi)
ax.set_ylim(-2, 2)
for i in range(2, 50):
   ax.plot(L, FuncSenFactorial(L, i)) 


#ax.plot(L, FuncSen(L, 2))
#ax.plot(L, sin(L))
```

```python
def FuncSen(x, orden):
    n=2
    term=x
    total=term
    for n in range(orden):
        term*=-x**2/((2*n-1)*(2*n-2))
        total+=term
    return total
```
