```python
%pylab inline
from matplotlib import animation
from scipy.optimize import curve_fit
```

```python
#Se extraen los datos del archivo .csv
data=genfromtxt("Rayleigh.csv",delimiter=",")
x=data[:,0]
y=data[:,1]
vx=data[:,2]
vy=data[:,3]
```

```python
#Se calcula un arreglo con las rapideces a partir de las componentes de velocidad
spd=np.sqrt(vx**2 + vy**2)
```

```python
#Se crean los arreglos de abscisas
t=np.linspace(0, 4, 100)
tnorm=np.linspace(-4, 4, 100)
#Se definen las funciones para hacer los ajustes a las distribuciones de Rayleigh (distr) y Gauss (normal)
def distr(t, sigma):
    return (t/sigma**2)*np.exp((-t**2)/(2*sigma**2))
def normal(t, a, b, c):
    return a*np.exp(-(b*(t-c)**2))
```

#Introducción

La distribución de Rayleigh es una distribución continua que se presenta cuando se tienen dos componentes ortigonales de velocidad con distribuciones normales y su norma o la rapidez tiene ésta distribución.
Se cree que un conjunto de partículas que interactúan según el potencial de Lennard-Jones, al alcanzar el equilibrio, tienen rapideces que se distribuyen según la distribución de Rayeigh. 

El potencial de Lennard-Jones se define como:

V(r)=4e((sigma/r)^12-(sigma/r)^6)

Donde e es la profundidad del pozo de potencial, sigma es la distancia finita en la que el potencial entre partículas es cero y r la distancia entre partículas. El término de orden 12 describe la repulsión y el de orden 6 describe la atracción.
Según lo antes mencionado acerca de dicha distribución, las velocidades en `x` y `y` en el equilibrio deben presentar una distribución normal.


#Análisis de datos

A la izquierda la distribución de las velocidades en x y a la derecha en y.
Al hacer la distribución de las rapideces se obtiene la figura 2, donde se ve que el ajuste es muy similar a una distribución de Rayleigh, lo que va acorde a nuestra hipótesis.
Finalmente se realizó una animación que muestra la evolución del sistema con 10000 iteraciones.


```python
#Se crea la Figura número 1
Figura1, axes=subplots(1, 2, figsize=(10, 7))
#Se dibujan los histogramas y se extrae la información de las barras (abscisa y ordenada)
nx, yx, eldex=axes[0].hist(vx[-1600:], normed=True)
ny, yy, eldey= axes[1].hist(vy[-1600:], normed=True)
#Se ajustan los datos extraídos a la distribución Gaussiana.
fitX, covmatX=curve_fit(normal, yx[1:], nx)
fitY, covmatY=curve_fit(normal, yy[1:], ny)
#Se dibujan las curvas de ajuste.
axes[0].plot(tnorm, normal(tnorm, fitX[0], fitX[1], fitX[2]), lw=5)
axes[1].plot(tnorm, normal(tnorm, fitY[0], fitY[1], fitY[2]), lw=5)
```

```python
#Se crea la Figura número 2-
Figura2, ax=subplots(1, 1, figsize=(10, 7))
#Se dibuja el histograma y se extrae la información acerca de las abscisas y ordenadas de cada barra.
n, bins, histograma=ax.hist(spd[-1600:], bins=20, normed=True)
#Se ajustan los datos a la función de distribución de Rayleigh.
fitR, covmatR =curve_fit(distr, bins[1:], n)
#Se dibuja la curva que se obtiene.
ax.plot(t, distr(t, fitR[0]), lw=5)
```
