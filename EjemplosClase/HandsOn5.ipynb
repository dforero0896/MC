#Lissajous

```python
t=np.linspace(-pi, pi, 1000)
Lisa, axes=subplots(5, 5, figsize=(10, 10))
def X(a, t):
    return np.cos(a*t)
def Y(b, t):
    return np.cos(b*t)
for i in range(5):
    for j in range(5):
        
        a=np.random.randint(1, 100)
        b=np.random.randint(1, 100)
        #print a, b
        axes[i, j].set_xlim(-1, 1)
        axes[i, j].set_ylim(-1, 1)
        axes[i, j].plot(X(a, t), Y(b, t))
        axes[i, j].axis("off")
Lisa.savefig("Lissajous.png")
```
