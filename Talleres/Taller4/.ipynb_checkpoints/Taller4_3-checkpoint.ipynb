```python
data = loadtxt("Taller4.csv", delimiter=',', dtype=None, skiprows=1)
x=data[0:, 0]
y=data[0:, 1]
print len(x)
t=np.linspace(0.138490669327, 4.45641816306, 24)
```

```python
sCubico=interpolate.interp1d(x, y, kind='cubic')
```
