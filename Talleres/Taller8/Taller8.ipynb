```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#%matplotlib inline
```

```python
Nt=3000
Nl=100
L=10. # Lado del cubo
tmin=0
tmax=.3
xmin=-L/2
xmax=L/2
ymin=-L/2
ymax=L/2
dt=(tmax-tmin)/Nt
dx=(xmax-xmin)/Nl
dy=(ymax-ymin)/Nl

v=425. # m/s
xcoords=np.linspace(xmin,xmax,Nl)
ycoords=np.linspace(xmin,xmax,Nl)
xmesh,ymesh=np.meshgrid(xcoords,ycoords)
tcoords=np.linspace(tmin,tmax,Nt)
r=v*dt/dx
#En dos caras opuestas:
#psi(x,y,0) = np.exp(-xmesh**2)*np.exp(-ymesh**2)
```

```python
drum=np.zeros((Nt,Nl,Nl))
# fijar la condición inicial
stdev=0.1
drum[0]=np.exp(-(xmesh)**2/(2*stdev**2))*np.exp(-(ymesh)**2/(2*stdev**2))
drum[1]=drum[0]+r**2/2.*(np.roll(drum[0],1,axis=0)+np.roll(drum[0],-1,axis=0)+np.roll(drum[0],1,axis=1)+np.roll(drum[0],-1,axis=1)-4*drum[0])
# Mostrar la condición inicial
plt.figure(figsize=(10,3))
plt.imshow(drum[0],cmap='gist_heat_r')
plt.show()
```

```python
for i in range(2,Nt):
    pdrum=drum[i-1]
    ppdrum=drum[i-2]
    extendido=np.zeros([Nl+2, Nl+2])
    extendidop=np.copy(extendido)
    extendidopp=np.copy(extendido)
    extendidop[1:-1, 1:-1]=pdrum
    extendidopp[1:-1, 1:-1]=ppdrum
    
    extendidop[1:-1, -1]=pdrum[-1, :]
    extendidop[-1, 1:-1]=pdrum[:, -1]
    extendidop[1:-1, 0]=pdrum[0, :]
    extendidop[0, 1:-1]=pdrum[:, 0]

    extendidopp[1:-1, -1]=ppdrum[-1, :]
    extendidopp[-1, 1:-1]=ppdrum[:, -1]
    extendidopp[1:-1, 0]=ppdrum[0, :]
    extendidopp[0, 1:-1]=ppdrum[:, 0]
    
    extendido=2*(1.-2.*r**2)*extendidop-extendidopp+r**2*(np.roll(extendidop,1,axis=1)+np.roll(extendidop,-1,axis=1)+np.roll(extendidop,1,axis=0)+np.roll(extendidop,-1,axis=0))
    drum[i]=extendido[1:-1, 1:-1]
```

```python
# Animación
from matplotlib import animation

fig=plt.figure(figsize=(10,10))
plt.hold(True)
plt.pcolor(xcoords,-ycoords,drum[0],cmap='coolwarm',vmin=0.,vmax=1.)
plt.xlim(xmin+dx,xmax-dx)
plt.ylim(ymin+dy,ymax-dy)
plt.colorbar()

def animate(i):
    plt.pcolor(xcoords,-ycoords,10*drum[5*i],cmap='coolwarm',vmin=0.,vmax=1.)
    plt.xlim(xmin,xmax)
    plt.ylim(ymin,ymax)
animacion = animation.FuncAnimation(fig, animate, np.arange(1,len(drum)/5),interval=5, blit=False)
plt.show()
plt.hold(False)
```

```python
cara1=np.zeros((Nt,Nl,Nl))
cara2=np.zeros((Nt,Nl,Nl))
cara3=np.zeros((Nt,Nl,Nl))
cara4=np.zeros((Nt,Nl,Nl))
cara5=np.zeros((Nt,Nl,Nl))
cara6=np.zeros((Nt,Nl,Nl))
# fijar la condición inicial
stdev=0.1
cara1[0]=np.exp(-(xmesh)**2/(2*stdev**2))*np.exp(-(ymesh)**2/(2*stdev**2))
cara1[1]=drum[0]+r**2/2.*(np.roll(drum[0],1,axis=0)+np.roll(drum[0],-1,axis=0)+np.roll(drum[0],1,axis=1)+np.roll(drum[0],-1,axis=1)-4*drum[0])

cara3[0]=np.exp(-(xmesh)**2/(2*stdev**2))*np.exp(-(ymesh)**2/(2*stdev**2))
cara3[1]=drum[0]+r**2/2.*(np.roll(drum[0],1,axis=0)+np.roll(drum[0],-1,axis=0)+np.roll(drum[0],1,axis=1)+np.roll(drum[0],-1,axis=1)-4*drum[0])
# Mostrar la condición inicial
plt.figure(figsize=(10,3))
plt.imshow(cara3[0],cmap='gist_heat_r')
plt.show()
```

```python
#Se define una función que deja lista la cara para ser actualizada.
def resCara(cara, caraArriba, dondeAr, caraAbajo, dondeAb, caraDerecha, dondeDer, caraIzquierda, dondeIzq):
    pdrum=cara[i-1]
    ppdrum=cara[i-2]
    caraArriba=caraArriba[i-1]
    caraAbajo=caraAbajo[i-1]
    caraDerecha=caraDerecha[i-1]
    caraIzquierda=caraIzquierda[i-1]
       
    if dondeAr == "derecha":
        caraArriba=caraArriba[:, -1]
    if dondeAr == "izquierda":
        caraArriba=caraArriba[:, 0]
    if dondeAr == "arriba":
        caraArriba=caraArriba[0, :]
    if dondeAr == "abajo":
        caraArriba=caraArriba[0, :]
        
    if dondeAb == "derecha":
        caraAbajo=caraAbajo[:, -1]
    if dondeAb == "izquierda":
        caraAbajo=caraAbajo[:, 0]
    if dondeAb == "arriba":
        caraAbajo=caraAbajo[0, :]
    if dondeAb == "abajo":
        caraAbajo=caraAbajo[0, :]
        
    if dondeDer == "derecha":
        caraDerecha=caraDerecha[:, -1]
    if dondeDer == "izquierda":
        caraDerecha=caraDerecha[:, 0]
    if dondeDer == "arriba":
        caraDerecha=caraDerecha[0, :]
    if dondeDer == "abajo":
        caraDerecha=caraDerecha[0, :]
        
    if dondeIzq == "derecha":
        caraIzquierda=caraIzquierda[:, -1]
    if dondeIzq == "izquierda":
        caraIzquierda=caraIzquierda[:, 0]
    if dondeIzq == "arriba":
        caraIzquierda=caraIzquierda[0, :]
    if dondeIzq == "abajo":
        caraIzquierda=caraIzquierda[0, :]

    #Se extiende.
    extendido=np.zeros([52, 52])
    extendidop=np.copy(extendido)
    extendidopp=np.copy(extendido)
    #Se define en el extendido.
    extendidop[1:-1, 1:-1]=pdrum
    extendidopp[1:-1, 1:-1]=ppdrum
    #Columna derecha.
    extendidop[1:-1, -1]=caraDerecha
    #Fila inferior.
    extendidop[-1, 1:-1]=caraAbajo
    #Columna izquierda
    extendidop[1:-1, 0]=caraIzquierda
    #Fila superior.
    extendidop[0, 1:-1]=caraArriba
    #Columna derecha.
    extendidopp[1:-1, -1]=caraDerecha
    #Fila inferior.
    extendidopp[-1, 1:-1]=caraAbajo
    #Columna izquierda.
    extendidopp[1:-1, 0]=caraIzquierda
    #Fila superior.
    extendidopp[0, 1:-1]=caraArriba
    
            
    extendido=2*(1.-2.*r**2)*extendidop-extendidopp+r**2*(np.roll(extendidop,1,axis=1)+np.roll(extendidop,-1,axis=1)+np.roll(extendidop,1,axis=0)+np.roll(extendidop,-1,axis=0))
    cara[i]=extendido[1:-1, 1:-1]
```

```python
for i in range(2,Nt):
    resCara(cara1, cara5, "abajo", cara6, "arriba", cara2, "izquierda", cara4, "derecha")
    resCara(cara2, cara4, "derecha", cara5, "derecha", cara3, "izquierda", cara1, "derecha")
    resCara(cara3, cara5, "arriba", cara6, "abajo", cara4, "izquierda", cara2, "derecha")
    resCara(cara4, cara5, "izquierda", cara6, "izquierda", cara1, "izquierda", cara3, "derecha")
    resCara(cara5, cara3, "arriba", cara1, "arriba", cara2, "arriba", cara4, "arriba")
    resCara(cara6, cara1, "abajo", cara3, "abajo", cara4, "abajo", cara2, "abajo")
```
